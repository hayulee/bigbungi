```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
plt.rc('font', family = 'Malgun Gothic')
```

```python
kids = pd.read_csv('kids_accident_Data.csv', engine='python')
zone = pd.read_csv('schoolzone_accident_Data.csv', engine='python')
```

```python
kids['tag']='어린이사고(14세이하)'
kids.head(5)
```

```python
zone['tag']='스쿨존사고(12세이하)'
zone.head(5)
```

```python
accident = kids.append(zone)
```

```python
accident.reset_index(drop=True)
```

```python
accident
```

```python
plt.figure(figsize=(50, 8))
sns.catplot(data = accident, x = '년도', y = '사고건수(건)', hue = 'tag', palette={"어린이사고(14세이하)": "g", "스쿨존사고(12세이하)": "m"},
            markers=["^", "o"], linestyles=["-", "--"], kind="point")
plt.title("년도별 어린이사고")
plt.ylabel("사고건수")
plt.show()

#  palette={"male": "g", "female": "m"},
            
# plt.bar(y_pos, height, color=['black', 'red', 'green', 'blue', 'cyan'])
# plt.xticks(y_pos, bars)
```

```python
plt.figure(figsize=(50, 8))
sns.catplot(data = zone, x = '년도', y = '사고건수(건)', hue = 'tag', palette={"스쿨존사고(12세이하)": "m"},
            markers="o", linestyles="-", kind="point")
plt.title("년도별 스쿨존사고")
plt.ylabel("사고건수")
plt.show()
```
